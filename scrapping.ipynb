# Scraping Dataset

```python
from google_play_scraper import app, reviews_all, Sort

# Mengambil ulasan dari aplikasi Gojek (ID: com.gojek.app)
scrapreview = reviews(
    'com.gojek.app',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count=15000  # Ini akan benar-benar berhenti di 11.000
)
```

```python
# Menyimpan ulasan dalam file CSV
import csv

with open('ulasan_aplikasi.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])  # Menulis header kolom
    for review in scrapreview[0]:  # Mengakses elemen pertama dari tuple scrapreview yang berisi daftar ulasan
        writer.writerow([review['content']])  # Menulis konten ulasan ke dalam file CSV
```
